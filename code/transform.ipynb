```python
df = pd.read_csv("../landing-zone/city_day.csv")

df['PM2.5'].replace([0, 0.0], df['PM2.5'].median(), inplace=True)
df['PM10'].replace([0, 0.0], df['PM10'].median(), inplace=True)
df['NO'].replace([0, 0.0], df['NO'].median(), inplace=True)
df['NO2'].replace([0, 0.0], df['NO2'].median(), inplace=True)
df['NOx'].replace([0, 0.0], df['NOx'].median(), inplace=True)
df['NH3'].replace([0, 0.0], df['NH3'].median(), inplace=True)
df['CO'].replace([0, 0.0], df['CO'].median(), inplace=True)
df['SO2'].replace([0, 0.0], df['SO2'].median(), inplace=True)
df['O3'].replace([0, 0.0], df['O3'].median(), inplace=True)
df['Benzene'].replace([0, 0.0], df['Benzene'].median(), inplace=True)
df['Toluene'].replace([0, 0.0], df['Toluene'].median(), inplace=True)
df['Xylene'].replace([0, 0.0], df['Xylene'].median(), inplace=True)
df['AQI'].replace([0, 0.0], df['AQI'].median(), inplace=True)

df['PM2.5'].fillna(df['PM2.5'].median(), inplace=True)
df['PM10'].fillna(df['PM10'].median(), inplace=True)
df['NO'].fillna(df['NO'].median(), inplace=True)
df['NO2'].fillna(df['NO2'].median(), inplace=True)
df['NOx'].fillna(df['NOx'].median(), inplace=True)
df['NH3'].fillna(df['NH3'].median(), inplace=True)
df['CO'].fillna(df['CO'].median(), inplace=True)
df['SO2'].fillna(df['SO2'].median(), inplace=True)
df['O3'].fillna(df['O3'].median(), inplace=True)
df['Benzene'].fillna(df['Benzene'].median(), inplace=True)
df['Toluene'].fillna(df['Toluene'].median(), inplace=True)
df['Xylene'].fillna(df['Xylene'].median(), inplace=True)
df['AQI'].fillna(df['AQI'].median(), inplace=True)



moda = df['AQI_Bucket'].mode()[0]
df['AQI_Bucket'] = df['AQI_Bucket'].fillna(moda)

df.to_csv("../refined-zone/city_day_refined.csv")



df.to_parquet("../raw-zone/parquet/city_day.parquet", engine="pyarrow", index=False)

# # Detalles de conexi√≥n a PostgreSQL
# usuario = 'postgres'
# contrase√±a = '2585'
# host = 'localhost'
# puerto = '5432'
# base_de_datos = 'postgres'

# # Crear el engine de conexi√≥n para PostgreSQL
# engine = create_engine(f'postgresql+psycopg2://{usuario}:{contrase√±a}@{host}:{puerto}/{base_de_datos}')

# # Insertar el DataFrame en PostgreSQL
# df.to_sql('city_day', engine, if_exists='replace', index=False)

# print("Datos insertados exitosamente.")

df.head()
```
